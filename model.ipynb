```python
#create a csv file with all the features
import csv
path = 'C:/Users/rehan/Desktop/term-preterm-ehg-dataset-with-tocogram-1.0.0/'
with open('data.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["filename" ,"RMS_EHG_1", "log2_EHG_1", "perm_entropy_EHG_1", "svd_entropy_EHG_1", "spectral_entropy_EHG_1", "sample_entropy_EHG_1", 
    "Hjorth_mobility_EHG_1","hjorth_complexity_EHG_1","Approximate entropy_EHG_1","num_zero_cross_EHG_1", "petrosian_fd_EHG_1", "katz_fd_EHG_1", "higuchi_fd_EHG_1", "detrended_fluctuation_EHG_1"])
    for file in os.listdir(path):
        if file.endswith(".dat"):
            record = wfdb.rdrecord(path+file[:-4])
            df = pd.DataFrame(record.p_signal)
            df.columns = record.sig_name
            df.to_csv(path+file[:-4]+'.csv',header=record.sig_name, index=True)
            dfex=pd.read_csv(path+file[:-4]+'.csv')
            dfex.drop(dfex.columns[0], axis=1, inplace=True)
            dfex.drop(dfex.columns[-1], axis=1, inplace=True)
            dfex.drop(dfex.columns[-1], axis=1, inplace=True)
            x=dfex['EHG1'].to_numpy()
            rms = np.sqrt(np.mean(x**2))
            log2 = np.log2(rms)
            perm_entropy = ant.perm_entropy(x, normalize=True)
            svd_entropy = ant.svd_entropy(x, normalize=True)
            spectral_entropy = ant.spectral_entropy(x, sf=100, method='welch', normalize=True)
            sample_entropy = ant.sample_entropy(x)
            hjorth_params = ant.hjorth_params(x)
            num_zerocross = ant.num_zerocross(x)
            petrosian_fd = ant.petrosian_fd(x)
            katz_fd = ant.katz_fd(x)
            higuchi_fd = ant.higuchi_fd(x)
            approximate_entropy=ant.app_entropy(x, order=2, metric='chebyshev')
            detrended_fluctuation = ant.detrended_fluctuation(x)
            writer.writerow([file,rms, log2, perm_entropy, svd_entropy, spectral_entropy, sample_entropy, hjorth_params[0], hjorth_params[1],approximate_entropy, num_zerocross, petrosian_fd, katz_fd, higuchi_fd, detrended_fluctuation])
```
